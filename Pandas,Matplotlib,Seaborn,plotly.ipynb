```python
np.random.seed(20)
df=pd.DataFrame(np.random.randn(10,4),index=pd.date_range('2021-7-18',periods=10),columns=['a',"b",'c','d'])
```

```python
df["new_col_square"]=df['d'].apply(lambda x: x**2)
```

```python
df["new_col_square-d_log"]=df['d'].apply(lambda x: np.log(x))
```

```python
df
```

```python
def test(x):
    if x<0:
        return x
    else :
        return np.log(x)
```

```python
df["new_col_square-d_log_1"]=df['d'].apply(test)
```

```python
df
```
